```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load the dataset
data = pd.read_csv('Fraud.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Check for data types and null values
data.info()

# Statistical summary
data.describe()

# Check for class imbalance
data['isFraud'].value_counts()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Visualize outliers using boxplots for numeric columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=data.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.show()

# Calculate Q1 and Q3 for numeric columns only
Q1 = data.select_dtypes(include=['float64', 'int64']).quantile(0.25, numeric_only=True)
Q3 = data.select_dtypes(include=['float64', 'int64']).quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

# Filter out outliers for numeric columns only
data = data[~((data.select_dtypes(include=['float64', 'int64']) < (Q1 - 1.5 * IQR)) | 
              (data.select_dtypes(include=['float64', 'int64']) > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Calculate correlation matrix for numeric columns only
corr_matrix = data.select_dtypes(include=['float64', 'int64']).corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# Selecting only numeric features for VIF calculation
numeric_data = data.select_dtypes(include=['float64', 'int64'])
vif_data = pd.DataFrame()
vif_data["feature"] = numeric_data.columns
vif_data["VIF"] = [variance_inflation_factor(numeric_data.values, i) for i in range(len(numeric_data.columns))]

# Display VIF
print(vif_data)
```

```python
# Assume 'isFraud' is the target variable and drop 'nameOrig' and 'nameDest' as they are identifiers
target = 'isFraud'
features = data.columns.drop([target, 'nameOrig', 'nameDest'])
```

```python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.3, random_state=42)

# Model training
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Feature importance
importance = rf_model.feature_importances_
feature_importance = pd.DataFrame({'feature': features, 'importance': importance}).sort_values(by='importance', ascending=False)

# Display feature importance
print(feature_importance)

# # Visualizing feature importance
# plt.figure(figsize=(12, 6))
# sns.barplot(x='importance', y='feature', data=feature_importance)
# plt.title('Feature Importance')
# plt.show()
```
